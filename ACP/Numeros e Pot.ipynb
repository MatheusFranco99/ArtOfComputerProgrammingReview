# Numeros, Potências e Logaritmos

### 1 Inteiros: 
São os números "completos":

$..., -3, -2 , -1, 0, 1, 2, 3, ...$

    -> negativos, zero, e positivos
	

### 2 Racionais: 
São a razão(quociente) entre dois inteiros, p/q, onde q é positivo.
 
### 3 Reais:
Incluem os racionais, e são uma quantidade x, com uma extensão decimal:
$x = n + 0,d1d2d3....$

    -> onde di é um dígito entre 0 e 9, e a sequência de dígitos não termina em infinitos 9 (isso seria um erro de representação ou algo mais profundo? 0.99999 = 1 etc...)
 	
Essa representação significa que, dado um k:

$n + d1/10 + d2/100 ... dk/100^k <= x < n + d1/10 + d2/100 ... dk/100^k + 1/10^k$

"A adição de um dígito em qualquer casa decimal de x é maior que x, e a sequência que forma x é menor ou igual a x, se incompleta ou completa."

Exemplos de números reais que não são racionais são:

    pi -> circunf/diâmetro de um círculo
    phi -> 'golden ratio'

### 4 Complexos: 

Que são uma quantidade da forma z = x + yi, onde x e y são reais e i é uma quantidade especial que satisfaz i² = -1.
Chama-se x da parte real e y da parte imaginária de z.

O valor absoluto de z é definido como:

$|z| = \sqrt{x² + y²}$

O complemento/conjugado complexo(?) de z é z\ = x - iy, e que satisfaz a condição z x z\ = x² + y² = |z|²

### Intervalos

Dados dois números reais u e v, tal que u =< v, dizemos que:

O itervalo fechado [u, v] é o conjunto de números reais x tal que u =< x =< v

O intervalo aberto (u, v) é o conjunto de numeros reais x tal que u < x < v

Intervalos podem ser semi abertos também: (u,v] e [u,v)
O intervalo (-infinito,+infinito) é o conjunto dos reais
